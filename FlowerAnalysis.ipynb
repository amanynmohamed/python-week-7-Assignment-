```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({i: name for i, name in enumerate(iris.target_names)})

# Display the first 5 rows
df.head()
```

The .info() function was then used to check the data types and ensure there are no missing values in the dataset.

```python
import matplotlib.pyplot as plt

# Simulated data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [250, 300, 400, 350, 450, 500]

# Line plot
plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', color='green')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

This kind of chart is useful for identifying upward or downward trends across a time-based variable like months or years.

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Compute average petal length per species
avg_petal = df.groupby('species')['petal length (cm)'].mean()

# Bar chart
plt.figure(figsize=(6, 4))
avg_petal.plot(kind='bar', color='skyblue')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.xticks(ticks=[0, 1, 2], labels=iris.target_names, rotation=0)
plt.grid(axis='y')
plt.show()
```

### Bar Chart – Average Petal Length per Species

This bar chart displays the average petal length for each iris species.  
By grouping the data and calculating the mean, we can see that species 2 has the highest average petal length.

Bar charts are ideal for comparing values between categories, especially when you want to highlight differences in means or totals.

```python
# Histogram of sepal width
plt.figure(figsize=(6, 4))
plt.hist(df['sepal width (cm)'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

### Histogram – Distribution of Sepal Width

The histogram shows the distribution of the sepal width values across the dataset.  
Most values are centered around 3.0 cm, indicating a somewhat normal distribution.

Histograms are useful for understanding the shape and spread of a single numerical variable.

```python
# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'],
            c=df['species'], cmap='viridis', edgecolor='black')
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)
plt.show()
```

### Scatter Plot – Sepal Length vs. Petal Length

This scatter plot shows the relationship between sepal length and petal length.  
Each point is colored by species, allowing us to observe clustering patterns.  
There's a clear positive correlation between the two features.

Scatter plots are useful for identifying relationships or correlations between two continuous variables.
